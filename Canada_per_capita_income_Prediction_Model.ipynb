```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

importing the dataset

```python
path = "/content/drive/MyDrive/Datasets/canada.csv"
df = pd.read_csv(path)
df
```

```python
plt.xlabel('Years',color = 'red')
plt.ylabel('Income per Capita ($ in 100 thousand)')
plt.scatter(df.year,df.income,marker = '+',color = 'blue')
```

```python
reg = LinearRegression()
```

```python
reg.fit(df[['year']],df.income)
```

```python
reg.predict([[2024]])
```

```python
plt.scatter(df.year,df.income,marker = '+',color = 'blue')
plt.plot(df.year,reg.predict(df[['year']]),color = 'blue')
```
